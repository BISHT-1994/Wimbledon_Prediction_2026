# Wimbledon 2026 Predictions

This notebook contains analysis and predictions for Wimbledon 2026 tennis tournament.

## Overview
- **Tournament**: Wimbledon 2026
- **Categories**: ATP (Men's) & WTP (Women's)
- **Goal**: Predict match outcomes and tournament winners

## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import json

# Set up visualization style
sns.set_style('darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)

print('✅ Libraries imported successfully!')
print(f'Current date: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
```

## 2. ATP Predictions

### ATP Players Data

```python
# Sample ATP players data
atp_players = {
    'Player': ['Novak Djokovic', 'Carlos Alcaraz', 'Jannik Sinner', 'Daniil Medvedev', 'Alexander Zverev'],
    'Rank': [1, 2, 3, 4, 5],
    'Wins': [85, 78, 72, 68, 65],
    'Win_Percentage': [0.82, 0.79, 0.76, 0.73, 0.70],
    'Grass_Court_Titles': [5, 2, 1, 0, 0]
}

atp_df = pd.DataFrame(atp_players)
print('\n📊 ATP Top Players:')
print(atp_df)
print(f'\nTotal players: {len(atp_df)}')
```

### ATP Predictions Analysis

```python
# Calculate predictions based on rankings and past performance
atp_df['Prediction_Score'] = (atp_df['Win_Percentage'] * 0.6) + (atp_df['Grass_Court_Titles'] * 0.4)

# Visualize ATP rankings
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Wins comparison
axes[0].barh(atp_df['Player'], atp_df['Wins'], color='steelblue')
axes[0].set_xlabel('Number of Wins')
axes[0].set_title('ATP Players - Total Wins')
axes[0].invert_yaxis()

# Prediction scores
axes[1].barh(atp_df['Player'], atp_df['Prediction_Score'], color='darkgreen')
axes[1].set_xlabel('Prediction Score')
axes[1].set_title('ATP Players - Prediction Scores')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

print('\n🎾 ATP Prediction Scores:')
atp_predictions = atp_df[['Player', 'Rank', 'Prediction_Score']].sort_values('Prediction_Score', ascending=False)
print(atp_predictions)
```

## 3. WTP (Women's Tennis) Predictions

### WTP Players Data

```python
# Sample WTP players data
wtp_players = {
    'Player': ['Iga Swiatek', 'Aryna Sabalenka', 'Marketa Vondrousova', 'Elena Rybakina', 'Jessica Pegula'],
    'Rank': [1, 2, 3, 4, 5],
    'Wins': [79, 75, 68, 64, 61],
    'Win_Percentage': [0.80, 0.77, 0.74, 0.71, 0.68],
    'Grass_Court_Titles': [1, 0, 0, 1, 0]
}

wtp_df = pd.DataFrame(wtp_players)
print('\n📊 WTP Top Players:')
print(wtp_df)
print(f'\nTotal players: {len(wtp_df)}')
```

### WTP Predictions Analysis

```python
# Calculate WTP predictions
wtp_df['Prediction_Score'] = (wtp_df['Win_Percentage'] * 0.6) + (wtp_df['Grass_Court_Titles'] * 0.4)

# Visualize WTP rankings
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Wins comparison
axes[0].barh(wtp_df['Player'], wtp_df['Wins'], color='hotpink')
axes[0].set_xlabel('Number of Wins')
axes[0].set_title('WTP Players - Total Wins')
axes[0].invert_yaxis()

# Prediction scores
axes[1].barh(wtp_df['Player'], wtp_df['Prediction_Score'], color='darkviolet')
axes[1].set_xlabel('Prediction Score')
axes[1].set_title('WTP Players - Prediction Scores')
axes[1].invert_yaxis()

plt.tight_layout()
plt.show()

print('\n🎾 WTP Prediction Scores:')
wtp_predictions = wtp_df[['Player', 'Rank', 'Prediction_Score']].sort_values('Prediction_Score', ascending=False)
print(wtp_predictions)
```

## 4. Comparative Analysis

```python
# Combine ATP and WTP data
atp_df['Category'] = 'ATP'
wtp_df['Category'] = 'WTP'

combined_df = pd.concat([atp_df, wtp_df], ignore_index=True)

# Create comparative visualizations
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Win percentage by category
sns.boxplot(data=combined_df, x='Category', y='Win_Percentage', ax=axes[0], palette=['steelblue', 'hotpink'])
axes[0].set_title('Win Percentage Distribution by Category')
axes[0].set_ylabel('Win Percentage')

# Prediction score by category
sns.boxplot(data=combined_df, x='Category', y='Prediction_Score', ax=axes[1], palette=['darkgreen', 'darkviolet'])
axes[1].set_title('Prediction Score Distribution by Category')
axes[1].set_ylabel('Prediction Score')

plt.tight_layout()
plt.show()

print('\n📈 Comparative Statistics:')
print(combined_df.groupby('Category')[['Wins', 'Win_Percentage', 'Prediction_Score']].describe())
```

## 5. Tournament Predictions Summary

```python
# Generate predictions summary
print('\n' + '='*60)
print('🏆 WIMBLEDON 2026 PREDICTIONS SUMMARY 🏆')
print('='*60)

print('\n🎾 ATP FAVORITES (Top 3):')
atp_top3 = atp_df.nlargest(3, 'Prediction_Score')[['Player', 'Rank', 'Prediction_Score']]
for idx, (_, row) in enumerate(atp_top3.iterrows(), 1):
    print(f"{idx}. {row['Player']} (Rank #{row['Rank']}) - Score: {row['Prediction_Score']:.3f}")

print('\n🎾 WTP FAVORITES (Top 3):')
wtp_top3 = wtp_df.nlargest(3, 'Prediction_Score')[['Player', 'Rank', 'Prediction_Score']]
for idx, (_, row) in enumerate(wtp_top3.iterrows(), 1):
    print(f"{idx}. {row['Player']} (Rank #{row['Rank']}) - Score: {row['Prediction_Score']:.3f}")

print('\n' + '='*60)
```
